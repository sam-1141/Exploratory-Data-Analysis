```python
path_country = "/content/drive/MyDrive/Data_Sets/Country-Code.xlsx"
df_country = pd.read_excel(path_country)
```

```python
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on = 'Country Code', how = 'left')
```

```python
final_df.describe()
```

```python
Country_Orders = final_df["Country"].value_counts().values
```

```python
Country_Orders
```

```python
Country_Names = final_df["Country"].value_counts().index
```

```python
Country_Names
```

### Let's Create a Pie-Chart

```python
plt.pie(Country_Orders[:3],labels=Country_Names[:3],autopct = "%1.2f%%")
```

```python
grp_by=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating_counts'})
```

```python
final_df['Rating color'].value_counts()
```

```python
plt.pie(final_df['Rating color'].value_counts().values,labels=final_df['Rating color'].value_counts().index, autopct="%1.2f%%")
```

```python
grp_by
```

```python
matplotlib.rcParams['figure.figsize'] = (12,6)
```

```python
final_df['Rating color'].value_counts().index
```

```python
pallete = ['White','Red','Orange','Yellow','Green','Green']
```

```python
sns.barplot(x='Aggregate rating', y='Rating_counts',hue = 'Rating color', data = grp_by, palette = pallete)
```
